```python
import os
import cv2
import pytesseract
import yt_dlp
```

```python
# Function to download the YouTube video using yt-dlp
def download_video_yt_dlp(youtube_url):
    try:
        ydl_opts = {
            'format': 'mp4',
            'outtmpl': 'downloaded_video.mp4'  # Save the video with this name
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([youtube_url])
        return 'downloaded_video.mp4'
    except Exception as e:
        print(f"Error occurred while downloading video using yt-dlp: {e}")
        return None

# Function to extract frames from the video and perform OCR
def extract_text_from_video(video_path, interval=5):
    if not os.path.exists('frames'):
        os.makedirs('frames')

    video = cv2.VideoCapture(video_path)
    frame_rate = video.get(cv2.CAP_PROP_FPS)
    frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"Extracting text from {frame_count} frames at {frame_rate} FPS...")

    extracted_text = []
    success, image = video.read()
    count = 0

    while success:
        if count % (frame_rate * interval) == 0:
            frame_name = f'frames/frame{count}.png'
            cv2.imwrite(frame_name, image)

            text = pytesseract.image_to_string(image)
            extracted_text.append(text)
            print(f"Extracted text from frame {count}: {text}")

        success, image = video.read()
        count += 1

    video.release()
    return extracted_text

# Main function to run the script
def main():
    youtube_url = input("Enter the YouTube video URL: ")

    print("Downloading video using yt-dlp...")
    video_path = download_video_yt_dlp(youtube_url)
    if not video_path:
        print("Video download failed. Please try again with a different link.")
        return

    print(f"Video downloaded at {video_path}")

    print("Extracting text from the video...")
    text_from_video = extract_text_from_video(video_path, interval=5)

    # Save the extracted text to a file
    with open("extracted_text.txt", "w") as f:
        for text in text_from_video:
            f.write(text + "\n")

    print("Text extraction complete. Saved to 'extracted_text.txt'")

# Run the main function
main()
```
